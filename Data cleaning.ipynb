```python
#task 5.
def bag_of_words(tokens : list[str]) -> dict[str, int]:
    word_bag ={}
    for t in tokens:
        if t in word_bag:
            word_bag[t] += 1
        else:
            word_bag[t] = 1
    return word_bag


#Task 6.
def most_frequent_words(word_bag : dict[str, int], n : int) -> list[str]:
    from collections import Counter
    c = Counter(word_bag)
    most_frequent = c.most_common(n)
    keylist = [key for key, val in most_frequent]
    return keylist


#Task8.
def statistics(word_bag : dict[str, int]) -> dict[str, float]:
    keylist = list(word_bag)
    valuelist = list(word_bag.values())

    import statistics
    count = sum(valuelist)

    length = []
    for k in keylist:
        length.append(int(len(k)))

    products = []

    for v, l in zip(valuelist, length):
        products.append(v * l)

    print(products)
    average_length = (sum(products))/count
    median_length = statistics.median(length)
    average_frequency = statistics.mean(valuelist)
    median_frequency = statistics.median(valuelist)

    statkeys = ['count', 'average_length', 'average_frequency', 'median_length', 'median_frequency']
    statvalues = [count, average_length, average_frequency, median_length, median_frequency]
    statlist = dict(zip(statkeys, statvalues))

    return statlist

print(statistics({
             "A": 10,
             "AB": 9,
             "ABC": 8,
             "ABCD": 7,
             "ABCDE": 6,
             "ABCDEF": 5,
             "ABCDEFG": 4,
             "ABCDEFGH": 3,
             "ABCDEFGHI": 2,
             "ABCDEFGHIJ": 1,
         }))
```
