```python
from time import time
import json
import ast
from pymongo import MongoClient
from pprint import pprint
```

```python
client = MongoClient("localhost", 27017)
db = client.CitationDB
papers = db.papers
```

```python
query ={"abstract": {"$exists": 1},  "year": {"$lt":2011}}
#projections = {"id":1, "abstract":1, "references":1, 'n_citation': 1, "year":1}
cited = list(db.papers.find(query))
```

query ={"abstract": {"$exists": 1},"references": {"$exists":1, "$ne": []}, "venue": {"$exists":1, "$ne": ''} "year": {"$eq":2010}}
#projections = {"id":1, "abstract":1, "references":1, 'n_citation': 1, "year":1}
focal_db = list(db.col.find(query))


```python
with open("2010pool.json") as json_file:
    data = json.load(json_file)
    focal_db = list(data)
```

```python
id_list =[]
for paper in cited:
    id_list.append(paper['id'])

#len(id_list) 1198903
```

```python
focal_without_reference =[]
for paper in reversed(focal_db):
    for references in paper['references']:    
        if references not in id_list:
            focal_without_reference.append(paper)
                
```

```python
focal_with_reference = []
for paper in focal_db:
    if paper not in focal_without_reference:
        focal_with_reference.append(paper)
        
```

```python
id_list =[]
for paper in focal_db:
    id_list.append(paper['id'])

ref_list=[]
for paper in focal_db:
    ref_list.append(paper['references'])
    
diction = dict(zip(id_list, ref_list))
```

```python
import random
dictionary = diction
test_sample = dict(random.sample(dictionary.items(), 5000))
```

```python
focal_paper_list = list(test_sample.keys())
reference_paper_list = []

for ref_list in test_sample.values():
    for ref in ref_list:
        reference_paper_list.append(ref)
#print(reference_paper_list)
```

```python
no_ref =[]

for focal, ref in test_sample.items():
    if focal not in focal_with_reference:
        no_ref.append(focal)
            #print(focal, paper)     
        
len(no_ref)
```

```python
no_ref = list(dict.fromkeys(no_ref))
```

```python
len(no_ref)
```

```python
for paper in reversed(focal_paper_list):
    if paper in no_ref:
        focal_paper_list.remove(paper)
```

```python
len(focal_paper_list)
```

```python
focal_data = []
for file in cited: 
    if file['id'] in focal_paper_list:
        focal_data.append(file)
```

```python
ref_list_clean = []
for liste in reference_paper_list:
    for paper in liste:
        ref_list_clean.append(paper)
#print(ref_list_clean)
ref_list_clean2 = []

for i in ref_list_clean:
    if i not in ref_list_clean2:
        ref_list_clean2.append(i)
#len(ref_list_clean2)
```

```python
data=focal_data
for line in data:
    line['_id'] = ""

file = json.dumps(data)
f = open("focal_data.json","w")
f.write(file)
f.close()
```

```python
refer_data=[]
for file in cited:
    if file["id"] in ref_list_clean2:
        refer_data.append(file)
```

```python

data=refer_data
for line in data:
    line['_id'] = ""

file = json.dumps(data)
f = open("refer_data.json","w")
f.write(file)
f.close()
```
